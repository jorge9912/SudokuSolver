```python
import numpy as np

# this prepares the array of variables
def pre_process_sudoku(path) :
    X = []
    lines = open(path, "r")
    for i in range(9) :
        line = lines.readline()
        chars = line.split()
        city = ['label', 0, 0]
        row = []
        j = 0
        for char in chars:
            row.append(int(char))
            j += 1
        X.append(row)
    return X


# returns true if a num is in a row
#  false otherwise
def in_row(X,row,num): 
    for i in range(9): 
        if(X[row][i] == num): 
            return True
    return False

# returns true if a num is in a col
#  false otherwise
def in_col(X,col,num): 
    for i in range(9): 
        if(X[i][col] == num): 
            return True
    return False

# returns true if a num is in a section
#  false otherwise
def in_section(X,row,col,num): 
    for i in range(3): 
        for j in range(3): 
            if(X[i+row][j+col] == num): 
                return True
    return False

# checks if a num can be added to row and col
#   on a sudoku board
def is_safe(X,row,col,num): 
    a = in_row(X,row,num)
    b = in_col(X,col,num)
    c = in_section(X,row - row%3,col - col%3,num)
    return not a and not b and not c

# iterates through a sudoku board and picks out
#  the first 0 entry it finds
def get_var(X): 
    for row in range(9): 
        for col in range(9): 
            if(X[row][col]==0): 
                return [row, col] 
    return False

# takes in a sudoku board and returns a 3d boolean array
# where each ijk entry is true if k+1 can go in X[i][j]
def get_sudoku_domains(X):
    D = []
    for i in range(9):
        a = []
        for j in range(9):
            b = []
            for k in range(9):
                b.append(is_safe(X, i, j, k+1))
            a.append(b)
        D.append(a)
    return D
```

```python
var_count_A = 0

def sudoku_backtrack_A(X):
    # we us a global var_count because otherwise the valuse would be
    #   reset at the recursive step
    global var_count_A
    if(var_count_A >= 10000):
        # we give up in order to save time
        return False
    var = get_var(X)
    var_count_A += 1
    #if var is false, that means the board is full
    #  i.e. puzzle solved!
    if(not var): 
        return True 
    # in the case that the puzzle is not solved
    for value in range(1,10):
        if(is_safe(X, var[0], var[1], value)):
            # if we add var, and it doesnt violate any constraints 
            X[var[0]][var[1]] = value
            result = sudoku_backtrack_A(X)
            if(result): 
                return result
        # we failed reset and retry
        X[var[0]][var[1]] = 0
    return False 
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# this function runs sudoku_backtrack_A on all of the files

var_count_A = 0

def get_data_sudoku_A(path) :
    global var_count_A
    setNum = {}
    for i in range(1,72):
        setNum[str(i)] = 0
    first = os.listdir(path)
    for file in first:
        newPath = path + "/" +file
        second = os.listdir(newPath)
        l = len(newPath) + 1
        for file1 in second :
            finalPath = newPath + "/" + file1
            if(finalPath[l] != '.' ) :
                s = finalPath[16:18]
                if(s[1] == '/') :
                    s = s[0]
                sudoku_backtrack_A(pre_process_sudoku(finalPath))
                setNum[s] += var_count_A
                var_count_A = 0
    data = [0] * 71
    for i in setNum:
        data[int(i)-1] = setNum[i]/10.0 
    return data
```

```python
# this takes about 2 min to complete
dataA = get_data_sudoku_A("sudoku_problems")
```

```python
def analysis(data):
    plt.plot(data)
    plt.ylabel("Average # of Variables Checked")
    plt.xlabel("# of Squares Filled to start")
    plt.title("Variables Checked Vs. # of Squared filled to star")
    plt.show()
```

```python
var_count_B = 0

# this is the forward checking function
def sudoku_forward_checking(X, D, row, col, value, tf):
    for i in range(9):
        if (X[i][col] == 0):
            D[i][col][value-1] = tf
        if (X[row][i] == 0):
            D[row][i][value-1] = tf
    for i in range(3):
        a = row - row%3 + i
        for j in range(3):
            b = col - col%3 + j
            if(X[a][b] == 0) :
                D[a][b][value-1] = tf
    return D

# we need to check if there is a spot where nothing
#  can go, in otherwords, we need to determine weather or
#  not the current puzzle is solvable
def is_viable(X, D):   
    for i in range(9):
        for j in range(9):
            if(X[i][j] == 0):
                # if we find a square that is empty
                #   and nothing can fit, we know the board
                #   is unsolvable in the current state
                for k in range(1,10):
                    if(D[i][j][k-1] == False):
                        break
                    if(k == 10):
                        return False
    return True
                        
    
def sudoku_backtrack_B(X, D):
    global var_count_B
    if(var_count_B >= 10000):
        #print("i give up")
        return False
    var = get_var(X)
    var_count_B += 1
    #if var is false, that means the board is full
    #  i.e. puzzle solved!
    if(not var): 
        return True 
    # in the case that the puzzle is not solved
    for value in range(1,10):
        if(D[var[0]][var[1]][value-1]): 
            # if we add var, and it doesnt violate any constraints 
            X[var[0]][var[1]] = value
            # use forward checking to update the domain.
            # i.e. upon addition of this value to X, what is now invalid?
            D = sudoku_forward_checking(X, D, var[0], var[1], value, False)
            if(is_viable(X, D)):
                result = sudoku_backtrack_B(X, D)
                if(result): 
                    return result
        # we failed reset and retry
        X[var[0]][var[1]] = 0
        D = sudoku_forward_checking(X, D, var[0], var[1], value, True)
    return False 
```

```python
# this function runs sudoku_backtrack_B on all of the files

var_count_B = 0

def get_data_sudoku_B(path) :
    global var_count_B
    setNum = {}
    for i in range(1,72):
        setNum[str(i)] = 0
    first = os.listdir(path)
    for file in first:
        newPath = path + "/" +file
        second = os.listdir(newPath)
        l = len(newPath) + 1
        for file1 in second :
            finalPath = newPath + "/" + file1
            if(finalPath[l] != '.' ) :
                s = finalPath[16:18]
                if(s[1] == '/') :
                    s = s[0]
                X = pre_process_sudoku(finalPath)
                D = get_sudoku_domains(X)
                sudoku_backtrack_B(X, D)
                setNum[s] += var_count_B
                var_count_B = 0
    data = [0] * 71
    for i in setNum:
        data[int(i)-1] = setNum[i]/10.0 
    return data
```

```python
dataB = get_data_sudoku_B("sudoku_problems")
```

```python
# the following version of get_var will select the square with
#  minimum remaining values heuristic
def get_var_C(X, D):
    best = 9
    return_r = 0
    return_c = 0
    for row in range(9):
        for col in range(9):
            size = 0
            for k in range(1,10):
                if(D[row][col][k-1] == True):
                    size += 1
            if(size < best) :
                return_r = row
                return_c = col
    return [return_r, return_c]

# this is the forward checking function that is modified to also retrun
#  the amount of squares that becomes unavialible upon addition of value
#  to the sudoku board
def sudoku_forward_checking_least_constraining(X, D, row, col, value, tf):
    count = 0
    for i in range(9):
        if (X[i][col] == 0):
            D[i][col][value-1] = tf
            count += 1
        if (X[row][i] == 0):
            D[row][i][value-1] = tf
            count += 1
    for i in range(3):
        a = row - row%3 + i
        for j in range(3):
            b = col - col%3 + j
            if(X[a][b] == 0) :
                D[a][b][value-1] = tf
                count += 1
    return count


var_count_C = 0

def sudoku_backtrack_C(X, D):
    global var_count_C
    if(var_count_C >= 10000):
        #print("i give up")
        return False
    
    var = get_var_C(X,D)
    var_count_C += 1
    #if var is false, that means the board is full
    #  i.e. puzzle solved!
    if(not var): 
        return True 
    # in the case that the puzzle is not solved
    arr = []
    count = 0
    # the job of the following loop is to create an array which contains the number
    #  of spaces that become unavailible when an assignment var = value is made
    for k in range(1,10):
        if(D[var[0]][var[1]][k-1]): 
            # if we add var, and it doesnt violate any constraints 
            X[var[0]][var[1]] = k
            arr.append(sudoku_forward_checking_least_constraining(X, D, var[0], var[1], k, False))
            count += 1
        else:
            arr.append(100)
    # this loop functions similarly to the loops we saw in A and B.  The difference is now
    #  instead of selecting value to be 1 to 10 in order, we are selecting value in [1,10]
    #  such that var = value is least constraining
    while(count >= 0):
        least_constraining = min(arr)
        value = arr.index(least_constraining) + 1
        arr[value] = arr[value-1] + 100
        D = sudoku_forward_checking(X, D, var[0], var[1], value, False)
        if(is_viable(X, D)):
            result = sudoku_backtrack_B(X, D)
            if(result): 
                return result
        # we failed reset and retry
        X[var[0]][var[1]] = 0
        D = sudoku_forward_checking(X, D, var[0], var[1], value, True)
        count -= 1
    return False
```

```python
# this function runs sudoku_backtrack_C on all of the files

var_count_C = 0

def get_data_sudoku_C(path) :
    global var_count_C
    setNum = {}
    for i in range(1,72):
        setNum[str(i)] = 0
    first = os.listdir(path)
    for file in first:
        newPath = path + "/" +file
        second = os.listdir(newPath)
        l = len(newPath) + 1
        for file1 in second :
            finalPath = newPath + "/" + file1
            if(finalPath[l] != '.' ) :
                s = finalPath[16:18]
                if(s[1] == '/') :
                    s = s[0]
                X = pre_process_sudoku(finalPath)
                D = get_sudoku_domains(X)
                sudoku_backtrack_C(X, D)
                setNum[s] += var_count_C
                var_count_C = 0
    data = [0] * 71
    for i in setNum:
        data[int(i)-1] = setNum[i]/10.0 
    return data
```
